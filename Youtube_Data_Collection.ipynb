```python
import requests
from bs4 import BeautifulSoup
import time
from selenium.webdriver.firefox.options import Options

options = Options()
#options.add_argument("--headless")
#options.add_argument('--window-size=1920,1080')
options.add_argument("--width=2560")
options.add_argument("--height=1440")

from selenium import webdriver
driver = webdriver.Firefox(options = options)


from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
```

```python
### Title and Links ###
```

```python
youtube_page = "https://www.youtube.com/@TheNewsBeak/videos"
data= []
video_links = []
wait = WebDriverWait(driver,15)
driver.get(youtube_page)

import random

for item in range(300):
        print("item",item)
        wait.until(EC.visibility_of_element_located((
            By.TAG_NAME, "body"))).send_keys(Keys.END)
        n = random.randint(4,8)
        time.sleep(n)


for video in driver.find_elements(
    By.CSS_SELECTOR, "#video-title-link"):
    d = []
    d.append(video.get_attribute("title"))
    d.append(video.get_attribute("href"))
    video_links.append(d)
    #print(len(video_links))
    
```

```python
import pandas as pd
df1 = pd.DataFrame(video_links, columns=['title','link'])
file_name = "video_links_2.csv"
df1.to_csv(file_name, encoding='utf-8', index=False)
```

```python
### Comments ###
```

```python

import random
video_comments = []
counter = 0
for urls in video_links:
    if (counter < 100):
        print("counter",counter)
        url = urls[1]
        scroll_arr = []
        data= []
        video_data = []
        video_data.append(url)
        
        options = Options()
        options.add_argument("--headless")
        options.add_argument("--width=2560")
        options.add_argument("--height=1440")

        driver = webdriver.Firefox(options = options)
        wait = WebDriverWait(driver,10)
        driver.get(url)

        scrollHeight = driver.execute_script("return window.scrollMaxY")
        scroll_arr.append(scrollHeight)



        for item in range(120):
            
                wait.until(EC.visibility_of_element_located(
                    (By.TAG_NAME, "body"))).send_keys(Keys.END)

                n = random.randint(4,6)
                time.sleep(n)
                prev_scroll_h = scroll_arr[-1]

                scrollHeight = driver.execute_script(
                    "return window.scrollMaxY")
                scroll_arr.append(scrollHeight)
                
            
                if (5 < len(scroll_arr) <= 14 
                    and scrollHeight == scroll_arr[-2]):
                    
                    wait.until(EC.visibility_of_element_located((
                        By.TAG_NAME, "body"))).send_keys(Keys.END)
                   
                    break
                    
                if (item > 15 and 
                    scrollHeight == prev_scroll_h):
                    
                    wait.until(EC.visibility_of_element_located((
                        By.TAG_NAME, "body"))).send_keys(Keys.END)
                    
                    break
                    
        try:

            for comment in wait.until(
                EC.presence_of_all_elements_located((
                    By.CSS_SELECTOR, "#content-text"))):
                    data.append(comment.text)

            video_data.append(data)
            counter+=1
            video_comments.append(video_data)
            driver.quit()
            
        except:
            pass
```

```python

df = pd.DataFrame(video_comments, columns=['url', 'comments'])
df.head()
file_name = "video_comments.csv"
df.to_csv(file_name, encoding='utf-8', index=False)
```
